```python
# %pip install segeval
from nltk import windowdiff, pk
from decimal import *
import segeval
import os
import xmlProsses
import numpy as np
import applcation as segapp
import math
```

```python
pathList=[]

for root, dirs, files in os.walk("/workspace/pj/AMI/topics", topdown=False):
    for name in files:
        pathList.append(name)
        
pathList=pathList[2:]
```

```python
def SLG(output, sList):
    SL = ''
    tmp =0
    SList = []
    lo = 0
    for s in sList:
        if (s not in output[lo]):
            lo+=1
            SL+='1'
            SList.append(tmp)
            tmp = 1
        else :
            SL+='0'
            tmp+=1
    SList.append(tmp)
    return SList
```

```python
score = []

for path in pathList:
    score.append([])
    result = xmlProsses.xmlProssesByPath(path)
    sList = [j for i in result[0] for j in i]
    for i in np.arange(0.25, 1.0, 0.025):
#     for i in np.arange(0.25, 1.0, 0.1):
        output = segapp.onlyTextProsess(sList, var=i)
        OSL = SLG(output, sList)
#         print(OSL, result[1])
        score[-1].append([segeval.window_diff(OSL, result[1]), segeval.pk(OSL, result[1]), segeval.boundary_similarity(OSL, result[1])])
        print(round(i, 3), 
              ' ', 
              [round(i, 3) for i in score[-1][-1]]
             )
    print('\n')
```

```python
avgMinWD = 0
avgMinPK = 0
avgMaxB = 0

for i in score:
    MinWD = 1
    MinPK = 1
    MaxB = 0
    
    for j in i:
#         print(j[0], j[1])
        if (MinWD > j[0]):
            MinWD = j[0]
        if (MinPK > j[1]):
            MinPK = j[1]
        if (MaxB < j[2]):
            MaxB = j[2]
            
    avgMinWD = avgMinWD + MinWD
    avgMinPK = avgMinPK + MinPK
    avgMaxB = avgMaxB + MaxB
    
avgMinWD = avgMinWD/len(score)
avgMinPK = avgMinPK/len(score)
avgMaxB = avgMaxB/len(score)
```
