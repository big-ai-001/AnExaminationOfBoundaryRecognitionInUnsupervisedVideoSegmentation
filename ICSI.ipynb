```python
from nltk import windowdiff, pk

import os
import xmlProssesICSI as xmlProsses
import numpy as np
import applcation as segapp
import math
```

```python
pathList=[]

for root, dirs, files in os.walk("/workspace/pj/ICSI_plus/Contributions/TopicSegmentation", topdown=False):
    for name in files:
        if ((not("Zone.Identifier" in name)) and (".topic.xml" in name)):
            pathList.append(name)
```

```python
pathList[14]
```

```python
def SLG(output, sList):
    SL = ''
    lo = 0
    for s in sList:
        if (s not in output[lo]):
            lo+=1
            SL+='1'
        else :
            SL+='0'
    return SL
```

```python
# pathList[14:]
```

```python
score = []

for path in pathList[14:]:
    try:
        score.append([])
        result, MSL = xmlProsses.xmlProssesByPath(path)
    except:
        continue
        
    sList = [j for i in result for j in i]
#     print(sList)
    for i in np.arange(0.0, 1.0, 0.025):
        output = segapp.onlyTextProsess(sList, var=i)
        OSL = SLG(output, sList)
        score[-1].append([windowdiff(MSL, OSL, k=3), pk(MSL, OSL)])
        print(i, ' ', score[-1][-1])
    print('\n')
```

```python
avgMinWD = 0
avgMinPK = 0

for i in score:
    MinWD = 1.0
    MinPK = 1.0
    
    for j in i:
#         print(j[0], j[1])
        if (MinWD > j[0]):
            MinWD = j[0]
        if (MinPK > j[1]):
            MinPK = j[1]
            
    avgMinWD += MinWD
    avgMinPK += MinPK
    
avgMinWD /= len(score)
avgMinPK /= len(score)
```
